# Detecting outliers and treating them with LTV and UTV

```python
# Detecting outliers in train dataset
def outlier_detect(titanic):
    for i in titanic.describe().columns:
        Q1=titanic.describe().at['25%',i]
        Q3=titanic.describe().at['75%',i]
        IQR=Q3-Q1
        LTV=Q1-1.5*IQR
        UTV=Q3+1.5*IQR
        titanic[i]=titanic[i].mask(titanic[i]<LTV,LTV) # Replacing outliers by LTV which are less than LTV
        titanic[i]=titanic[i].mask(titanic[i]>UTV,UTV) # Replacing outliers by UTV which are less than UTV
    return titanic
```

```python
titanic=outlier_detect(titanic)
```

## Hierarchical clustering

```python
# Import importatn libraries i.e. KMeans, dendrogram, linkage,AgglomerativeClustering, cdist
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
titanic1=titanic.drop(columns=['Name','Ticket']) # Dropping unnecessary columns
```

```python
titanic1
```

```python
data = titanic1.iloc[:, 1:6].values
# iloc[] is used when the index labels of a dataframe is something other than numeric series of 0,1,2,... or in case the user doesn't know the index labels.
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(30, 20))
plt.title("Train Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
# Dendrogram represents nested clusters. A dendrogram is a type of tree diagram showing hierarchical clustering relationship between sets of data.
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
# Applying Agglomerative clustering for 5 no. of clusters for prediction
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,3], c=cluster.labels_, cmap='rainbow')
# Displaying scatter plot of 'data' dataframe
```

## K-Means clustering

```python
# Import important library i.e. KMeans
from sklearn.cluster import KMeans
kmeans5 = KMeans(n_clusters=5)           # No. of clusters are 5
y_kmeans5 = kmeans5.fit_predict(titanic1)  #Applying kmeans5.fit_predict on train1 dataframe
print(y_kmeans5)

kmeans5.cluster_centers_
```

```python
# Finding no. of error values of specific range
Error =[]
# Applying for loop for 11 no. of loops to check error
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(titanic1)
    kmeans.fit(titanic1)
    Error.append(kmeans.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)       # Applying Elbow method to dispaly the range of error
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
# According to this plot, maximum no. of errors lies in no. of clusters=2 and extreme less error in no. of cluster=10
```

```python
plt.scatter(data[:,0],data[:,3],c=y_kmeans5, cmap='rainbow')
# Displaying scatter plot of 'data' dataframe
```
