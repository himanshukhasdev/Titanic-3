# Detecting outliers and treating them with LTV and UTV


```python
# Detecting outliers in titanic dataset
def outlier_detect(titanic):
    for i in titanic.describe().columns:
        Q1=titanic.describe().at['25%',i]
        Q3=titanic.describe().at['75%',i]
        IQR=Q3-Q1
        LTV=Q1-1.5*IQR
        UTV=Q3+1.5*IQR
        titanic[i]=titanic[i].mask(titanic[i]<LTV,LTV) # Replacing outliers by LTV which are less than LTV
        titanic[i]=titanic[i].mask(titanic[i]>UTV,UTV) # Replacing outliers by UTV which are less than UTV
    return titanic
```

```python
titanic=outlier_detect(titanic)
```

# Considering train and test variables

```python
titanic1=titanic.drop(columns=['Survived']) # Dropping Survival column as it is the output variable.
```

```python
x=titanic1         # Input is denoted by x
x
```

```python
y=titanic["Survived"]  # Output is denoted by y and it contains Survival column.
y
```

# Train and testing on i/p & o/p variables with 0.25(25%) test size


```python
# Import train_test_split using sklearn.model_selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
# Splitting the data in training and testing parts with test size 0.25 i.e. 25% of the overall data.
```

# Applying various classifiers as the output variable is discrete

## RandomForest Classifier

```python
# Fitting Random Forest Classification to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score 
Accuracy_Score = accuracy_score(y_test, y_pred)
Accuracy_Score                                   # Accuracy of random forest classifier
```

## DecisionTree Classifier

```python
# Fitting Decision Tree Classification to the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score 
Accuracy_Score = accuracy_score(y_test, y_pred)
Accuracy_Score                                    # Accuracy of decision tree classifier
```
